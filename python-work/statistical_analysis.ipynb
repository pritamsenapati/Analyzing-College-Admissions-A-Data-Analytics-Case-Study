GPA Hypothesis Test

```python
# Do admitted students have significantly higher GPAs
accepted = df[df["admission_status"]==1]["high_school_gpa"]

rejected = df[df["admission_status"]==0]["high_school_gpa"]
```

Independent T-Test

```python
t_stat, p_value = stats.ttest_ind(
    accepted,
    rejected,
    equal_var=False
)

print("T Statistic :", t_stat)

print("P Value :", p_value)
```

Interpretation

```python
if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

SAT Hypothesis Test

```python
# <!-- Interview Score Test -->
accepted = df[df["admission_status"]==1]["interview_score"]

rejected = df[df["admission_status"]==0]["interview_score"]

stats.ttest_ind(
    accepted,
    rejected,
    equal_var=False
)
```

```python
# Family Income Test
accepted = df[df["admission_status"]==1]["family_income"]

rejected = df[df["admission_status"]==0]["family_income"]

stats.ttest_ind(
    accepted,
    rejected,
    equal_var=False
)
```

```python
# GPA Confidence Interval
mean = accepted.mean()

std = accepted.std()

n = len(accepted)

ci = stats.t.interval(
    confidence=0.95,
    df=n-1,
    loc=mean,
    scale=std/np.sqrt(n)
)

print(ci)
```
