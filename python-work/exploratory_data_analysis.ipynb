```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("college_admission_clean.csv")
```

```python
df.head()
```

Univariate Analysis

```python
# What does the applicant pool look like
# GPA Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["high_school_gpa"], bins=30, kde=True)
plt.title("Distribution of High School GPA")
plt.show()
```

```python
# SAT Score Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["sat_score"], bins=30, kde=True)
plt.title("Distribution of SAT Scores")
plt.show()
```

```python
# ACT Score Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["act_score"], bins=30, kde=True)
plt.title("Distribution of ACT Scores")
plt.show()
```

```python
# Admission Status
sns.countplot(x="admission_status", data=df)
plt.title("Admission Status")
plt.show()
```

```python
# Gender Distribution
sns.countplot(x="gender", data=df)

plt.title("Gender Distribution")

plt.show()
```

Correlation Analysis

```python
# Which variables are most related to admission
plt.figure(figsize=(16,10))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
# GPA vs Admission
# Does GPA influence admission
sns.boxplot(
    x="admission_status",
    y="high_school_gpa",
    data=df
)

plt.title("GPA by Admission Status")

plt.show()
```

```python
# SAT vs Admission
sns.boxplot(
    x="admission_status",
    y="sat_score",
    data=df
)

plt.title("SAT Score by Admission")

plt.show()
```

```python
# ACT vs Admission
sns.boxplot(
    x="admission_status",
    y="act_score",
    data=df
)

plt.title("ACT Score by Admission")

plt.show()
```

```python
# Attendance
sns.boxplot(
    x="admission_status",
    y="attendance_rate",
    data=df
)

plt.title("Attendance Rate by Admission")

plt.show()
```

```python
# Student Activities
# Volunteer Hours
sns.boxplot(
    x="admission_status",
    y="volunteer_hours",
    data=df
)

plt.show()
```

```python
# Leadership
sns.boxplot(
    x="admission_status",
    y="leadership_positions",
    data=df
)

plt.show()
```

```python
# Coding Projects
sns.boxplot(
    x="admission_status",
    y="coding_projects",
    data=df
)

plt.show()
```

```python
# Certifications
sns.boxplot(
    x="admission_status",
    y="online_certifications",
    data=df
)

plt.show()
```

```python
# Personal Evaluation
sns.boxplot(
    x="admission_status",
    y="essay_score",
    data=df
)

plt.show()
```

State Analysis

```python
# Top applicant states
plt.figure(figsize=(14,6))

df["state"]\
.value_counts()\
.head(10)\
.plot(kind="bar")

plt.title("Top 10 States by Applicants")

plt.show()
```

Pairplot

```python
sample = df.sample(5000, random_state=42)

sns.pairplot(
    sample[
        [
            "high_school_gpa",
            "sat_score",
            "essay_score",
            "interview_score",
            "admission_status"
        ]
    ],
    hue="admission_status"
)

plt.show()
```
